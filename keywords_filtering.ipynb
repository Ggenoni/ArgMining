```python
useful_keywords = llama3_1["metrics"]["npw_pmi"]["label"]["Useful"]
useful_keywords
```

```python
useful_keywords_2 = llama3_2["metrics"]["npw_pmi"]["label"]["Useful"]
useful_keywords_2
```

```python
non_useful_keywords = llama3_1["metrics"]["npw_pmi"]["label"]["Not useful"]
non_useful_keywords
```

```python
non_useful_keywords_2 = llama3_2["metrics"]["npw_pmi"]["label"]["Not useful"]
non_useful_keywords_2
```

```python
print(f'Number of useful 1-gram tokens: {len(useful_keywords)}')
print(f'Number of useful 1-gram tokens: {len(non_useful_keywords)}')
```

```python
print(f'Number of useful 2-gram tokens: {len(useful_keywords_2)}')
print(f'Number of useful 2-gram tokens: {len(non_useful_keywords_2)}')
```

```python
matching_keywords = []

for keyword in non_useful_keywords.keys():
    if keyword in useful_keywords.keys():
        matching_keywords.append(keyword)

print(f"Number of matching useful and non useful keywords: {len(matching_keywords)}")
print(matching_keywords)
```

```python
matching_keywords_2 = []

for keyword in non_useful_keywords_2.keys():
    if keyword in useful_keywords_2.keys():
        matching_keywords_2.append(keyword)

print(f"Number of matching useful and non useful keywords: {len(matching_keywords_2)}")
print(matching_keywords_2)
```

```python
non_useful_keywords_2
```

```python
non_useful_scores = []

for keyword, score in non_useful_keywords.items():
    if keyword in matching_keywords:
      score = score - useful_keywords[keyword]
      non_useful_scores.append((keyword, score))
    else:
      non_useful_scores.append((keyword, score))

sorted_non_useful_scores = sorted(non_useful_scores, key=lambda x: x[1], reverse=True)
filtered_sorted_non_useful_scores = [x for x in sorted_non_useful_scores if x[1] > 0]

print(len(non_useful_scores))
print(non_useful_scores)
print(len(sorted_non_useful_scores))
print(sorted_non_useful_scores)
print(len(filtered_sorted_non_useful_scores))
print(filtered_sorted_non_useful_scores)
```

```python
non_useful_scores_2 = []

for keyword, score in non_useful_keywords_2.items():
    if keyword in matching_keywords_2:
      score = score - useful_keywords_2[keyword]
      non_useful_scores_2.append((keyword, score))
    else:
      non_useful_scores_2.append((keyword, score))

sorted_non_useful_scores_2 = sorted(non_useful_scores_2, key=lambda x: x[1], reverse=True)
filtered_sorted_non_useful_scores_2 = [x for x in sorted_non_useful_scores_2 if x[1] > 0]

print(len(non_useful_scores_2))
print(non_useful_scores_2)
print(len(sorted_non_useful_scores_2))
print(sorted_non_useful_scores_2)
print(len(filtered_sorted_non_useful_scores_2))
print(filtered_sorted_non_useful_scores_2)
```
